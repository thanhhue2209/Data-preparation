**Exercise 1:**

```python
import pandas as pd
import numpy as np
```

```python
nyc = pd.read_csv('nyc_weather_2018.csv')
nyc.head(5)
```

```python
sta = pd.read_csv('weather_stations.csv')
sta.head(5)
```

```python
new = nyc.merge(
    sta.rename(dict(id = 'station'), axis =1),
    on = 'station')
new.sample(5, random_state = 0)
```

```python
 snow = new[(new.datatype == 'SNOW') & (new.name.str.contains('NJ US'))]
 snow.drop_duplicates('station', inplace= True)
 snow.shape
```

```python
snow[(snow.datatype == "SNOW")&(snow.name.str.contains("NJ US"))]['elevation'].mean()
```

```python
e = pd.read_csv('earthquakes.csv')
e.head()
```

```python
ml = e[e.magType == 'ml']
e = e.assign(bin = pd.cut(ml.mag, bins = 5, \
            labels = ['nor', 'low', 'med', 'high', 'dangerous']))
e.bin.value_counts()
```

**Exercise 4:**

```python
wbs = pd.read_csv('weather_by_station.csv')
wbs.head()
```

```python
snow = wbs[wbs.datatype == 'SNOW']
snow.head()
```

```python
snow = snow.assign(month = pd.DatetimeIndex(snow.date).month)
snow.head()
```

```python
snow.groupby(['station', 'month']).mean()
```

```python
pd.crosstab(index = snow.station, columns = snow.month, colnames=['month'])
```
